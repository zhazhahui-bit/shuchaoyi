```python
import urllib.request
import json
import requests
from bs4 import BeautifulSoup
from requests.exceptions import RequestException
import re
```

```python
url = 'https://www.danke.com/room/gz?page=1'
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36"
headers = {'User-Agent': user_agent}
request = urllib.request.Request(url, headers = headers)
response = urllib.request.urlopen(request)
html = response.read().decode("utf-8")
soup = BeautifulSoup(html,"html.parser")
```

```python
list = soup.find(class_='r_ls_box').find_all(class_='r_lbx')
for item in list:
    room_id = item.find(class_='location').string
    room_price = item.find(class_='ty_b').string
    room_hezu = item.find(class_='r_lbx_cenb').find('i').string
    room_area = re.search(r'\d+',item.find(class_='r_lbx_cenb').text.replace(" ","").replace("\n","")).group(0)
    room_floor = re.search(r'\d+楼',item.find(class_='r_lbx_cenb').text.replace(" ","").replace("\n","")).group(0).replace("楼","")
    room_xiaoqu = item.find(class_='r_lbx_cena').find('a')['xiaoqu']
    room_position = item.find(class_='r_lbx_cena').find('a')['title'].split(" ")[0]
    room_type = item.find(class_='r_lbx_cena').find('a')['title'].split(" ")[3]
    print(room_id,room_price,room_hezu,room_area,room_floor,room_xiaoqu,room_position,room_type)
```

```python
import codecs
for i in range(1,3):
    url='https://www.danke.com/room/gz?page={}'.format(i)
    user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36"
    headers = {'User-Agent': user_agent}
    request = urllib.request.Request(url, headers = headers)
    response = urllib.request.urlopen(request)
    html = response.read().decode("utf-8")
    soup = BeautifulSoup(html,"html.parser")
    list1 = soup.find(class_='r_ls_box').find_all(class_='r_lbx')
    with codecs.open('蛋壳房源.csv','a+',encoding='utf-8') as file:
        for item in list1:
            room_id = (i-1)*20+int(item.find(class_='location').string)
            room_price =item.find(class_='ty_b').string
            room_hezu = item.find(class_='r_lbx_cenb').find('i').string 
            room_area = re.search(r'\d+',item.find(class_='r_lbx_cenb').text.replace(" ","").replace("\n","")).group(0)
            room_floor = re.search(r'\d+楼',item.find(class_='r_lbx_cenb').text.replace(" ","").replace("\n","")).group(0).replace("楼","")
            room_xiaoqu = item.find(class_='r_lbx_cena').find('a')['xiaoqu']
            room_position = item.find(class_='r_lbx_cena').find('a')['title'].split(" ")[0]
            room_type = item.find(class_='r_lbx_cena').find('a')['title'].split(" ")[3]
            room_info = [str(room_id),str(room_price),str(room_hezu),str(room_area),str(room_floor),str(room_xiaoqu),str(room_position),str(room_type)]
            print(room_info)
            str_write = ','.join(room_info)+'\n'
            file.write(str_write)
        
```
